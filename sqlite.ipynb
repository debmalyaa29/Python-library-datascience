```python
import sqlite3
```

```python
##connect to an SQLite database
connection=sqlite3.connect('example.db')
connection
```

```python
cursor=connection.cursor()
```

```python
##Create a Table
cursor.execute(''' 
Create Table if Not Exists employees(
id Integer Primary Key,
name Text Not Null,
age Integer,
department text
)
''')

##Commit the changes
connection.commit()
```

```python
cursor.execute("""
INSERT INTO employees (name, age, department)
VALUES
('Amit Sharma', 28, 'HR'),
('Neha Verma', 32, 'Finance'),
('Rahul Das', 25, 'IT'),
('Priya Singh', 29, 'Marketing'),
('Sourav Paul', 35, 'Operations')
""")

connection.commit()
```

```python
##Query the data from the tab;e
cursor.execute ('Select * from employees')
rows=cursor.fetchall()

##print the queried data
for row in rows:
    print(row)
```

```python
##Update data in the table
cursor.execute("""
UPDATE employees
SET age = 36
WHERE name = 'Sourav Paul'
""")

connection.commit()
```

```python
##working with student-data
connection=sqlite3.connect('student-dataset.db')
cursor=connection.cursor()

##create a trable for student data
cursor.execute("""
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    age INTEGER,
    gender TEXT,
    course TEXT,
    marks REAL
)
""")

##inserting data into the sales table
cursor.execute("""
INSERT INTO students (name, age, gender, course, marks)
VALUES
('Ayan Das', 20, 'Male', 'Data Science', 85.5),
('Riya Sen', 21, 'Female', 'Computer Science', 92.0),
('Sourav Paul', 22, 'Male', 'AI & ML', 78.0),
('Ananya Roy', 20, 'Female', 'Cyber Security', 88.5),
('Rahul Sharma', 23, 'Male', 'Computer Science', 81.0)
""")

data = [
    ('Amit Das',),
    ('Riya Gupta',),
    ('Sourav Sen',),
    ('Neha Roy',)
]

cursor.executemany("""
INSERT INTO students (name)
VALUES (?)
""", data)

connection.commit()


```

```python
cursor.execute("SELECT * FROM students")

rows = cursor.fetchall()
for row in rows:
    print(row)
```

```python
cursor.execute("SELECT name FROM students")

for row in cursor.fetchall():
    print(row[0])
```

```python
cursor.execute("SELECT id, name, course, marks FROM students")

for id_, name, course, marks in cursor.fetchall():
    print(f"ID: {id_} | Name: {name} | Course: {course} | Marks: {marks}")
```
