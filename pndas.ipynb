```python
##Accessing data from dataframe
##Create a Data Frame from a List of Dictionaries
import pandas as pd

data = [
    {"Name": "Alice", "Age": 25, "City": "Delhi"},
    {"Name": "Bob", "Age": 30, "City": "Mumbai"},
    {"Name": "Charlie", "Age": 35, "City": "Kolkata"}
]

df = pd.DataFrame(data)
df['Name']
df.iloc[0]
```

```python
##Accessing a specific element
print(df.at[1,'Age'])
```

```python
##Accessing a specified element using iat
df.iat[2,2]
```

```python
##Data Manupulation with DataFrames
df['Salary']=[50000,60000,70000] ##adding a column
print(df)
```

```python
##Remove a column
df.drop('Salary',axis=1,inplace =True)## 1 for colum and 0 for row
##inplace= True make the operation permanent
```

```python
##Add age to column
df['Age']=df['Age']+1
df
```

```python
##Display the datatype of eachcolumns
print("Data types:",df.dtypes)
##Describe the DataFramess
print(df.describe)
##Group by a colum and perform a aggretion
result = df.groupby("City")["Salary"].mean()
print(result)
```
