```python
##filling missing values with the mean of the colium
##let year have missing values
import pandas as pd
df=pd.read_csv('student-dataset.csv')
df['ethic_new']=df['ethnic.group'].fillna(df['ethnic.group'].mean())
df
```

```python
##Renaming Columns
df = df.rename(columns={
    'math.grade': 'math_grade'
})

df.head(5)
```

```python
##change datatype
df['age_new']=df['age'].astype(float)
df.head(2)
```

```python
## data manupulation
df['age_new']=df['age'].apply(lambda x:x*2)
df.head()
```

```python
##Data Aggregationg and Grouping
grouped_sum = df.groupby(['gender', 'nationality'])['math_grade'].sum()
print(grouped_sum)
```

```python
##Aggregate multiple function
df.groupby(['gender', 'nationality']).agg({
    'age': ['mean', 'max'],
    'math_grade': ['mean'],
    'english.grade': ['mean'],
    'id': ['count']
})
```

```python
##Merging and joining dataframes
df1 = pd.DataFrame({
    'student_id': [1, 2, 3, 4],
    'name': ['Kiana', 'Rahul', 'Emma', 'John'],
    'gender': ['F', 'M', 'F', 'M'],
    'math_grade': [3.7, 3.2, 3.9, 3.5]
})
df2 = pd.DataFrame({
    'student_id': [1, 2, 4, 5],
    'nationality': ['China', 'India', 'USA', 'UK'],
    'age': [22, 23, 25, 24]
})

df1
```

```python
pd.merge(df1,df2,on="student_id",how="inner")##common student_id will be consider
```

```python
pd.merge(df1,df2,on="student_id",how="outer")## both the student_id will be consider
```

```python
pd.merge(df1,df2,on="student_id",how="left")
```

```python
pd.merge(df1,df2,on="student_id",how="right")
```
